```python
#✅ Q1: Demonstrate 3 different methods for creating identical 2D arrays in NumPy

import numpy as np

# Method 1: Using np.array()
a1 = np.array([[1, 2], [3, 4]])

# Method 2: Using np.full()
a2 = np.full((2, 2), [[1, 2], [3, 4]])[:2]  # slicing for same shape

# Method 3: Using reshaped np.arange()
a3 = np.arange(1, 5).reshape(2, 2)

print(a1)
print(a2)
print(a3)
```

```python
# ✅ Q3: Explain:
# (a) Difference in np.array vs np.asarray

# np.array always makes a copy
x = [1, 2, 3]
arr1 = np.array(x)

# np.asarray returns the same object if already ndarray
arr2 = np.asarray(arr1)

print("Is copy created by array?", arr1 is not x)        # True
print("Is same object for asarray?", arr2 is arr1)
```

```python
#(b) Deep Copy vs Shallow Copy


# Shallow copy
a = np.array([1, 2, 3])
b = a  # No copy

# Deep copy
c = a.copy()  # Independent copy

a[0] = 100
print("Shallow Copy affected:", b)  # b[0] = 100
print("Deep Copy unaffected:", c)   # c[0] = 1
```

```python
#✅ Q5: Create NumPy array with integers (1–10) of shape (5, 6), then:
# (a) Extract all even integers
# (b) Subtract odd integers

arr = np.random.randint(1, 11, size=(5, 6))
even_nums = arr[arr % 2 == 0]
odd_subtracted = np.where(arr % 2 != 0, arr - 1, arr)

print("Original:\n", arr)
print("Even Integers:\n", even_nums)
print("Odd Subtracted:\n", odd_subtracted)
```

```python


# ✅ Q6: Create a 3D NumPy array (3, 3, 3) with random ints between 1–10 and:
# (a) Find max index in each 2D slice (axis=2)
# (b) Perform element-wise multiplication between array and its index max values

arr = np.random.randint(1, 11, size=(3, 3, 3))
max_indices = np.argmax(arr, axis=2)

print("3D Array:\n", arr)
print("Max Index (axis=2):\n", max_indices)

# For (b), reshape max indices to multiply across the right axis
multiplied = arr * max_indices[:, :, np.newaxis]
print("Element-wise Multiplication:\n", multiplied)
```

```python
# Q13: Create three NumPy arrays
x = np.random.rand(100)
y = np.random.rand(100)
z = np.random.rand(100)

import matplotlib.pyplot as plt

# a) Red circle marker scatter plot
plt.scatter(x, y, color='red', marker='o')
plt.xlabel('x-axis')
plt.ylabel('y-axis')
plt.title('Q3: Scatter Plot x vs y')
plt.grid(True)
plt.show()

# b) Add horizontal line at y = 0.5
plt.scatter(x, y, color='red', marker='o')
plt.axhline(y=0.5, color='blue', linestyle='--')
plt.xlabel('x-axis')
plt.ylabel('y-axis')
plt.title('Q3: Scatter with y = 0.5 line')
plt.grid(True)
plt.show()

# c) Add vertical line at x = 0.5
plt.scatter(x, y, color='red', marker='o')
plt.axvline(x=0.5, color='green', linestyle='--')
plt.xlabel('x-axis')
plt.ylabel('y-axis')
plt.title('Q3: Scatter with x = 0.5 line')
plt.grid(True)
plt.show()
```

```python

# Q14: Create time-series data and dual y-axis plot
import pandas as pd
import matplotlib.pyplot as plt

dates = pd.date_range(start='2023-01-01', periods=50)
temp = np.random.randint(20, 35, size=50)
humidity = np.random.randint(40, 80, size=50)

df_weather = pd.DataFrame({'Date': dates, 'Temperature': temp, 'Humidity': humidity})

fig, ax1 = plt.subplots()
ax2 = ax1.twinx()

ax1.plot(df_weather['Date'], df_weather['Temperature'], 'g-')
ax2.plot(df_weather['Date'], df_weather['Humidity'], 'b-')

ax1.set_xlabel('Date')
ax1.set_ylabel('Temperature', color='g')
ax2.set_ylabel('Humidity', color='b')
plt.title('Q4: Temperature and Humidity Over Time')
plt.show()
```

```python


# Q15: NumPy histogram and PDF overlay
samples = np.random.normal(loc=0, scale=1, size=1000)

plt.hist(samples, bins=30, density=True, alpha=0.6, color='c', label='Histogram')
x_vals = np.linspace(-4, 4, 1000)
pdf_vals = (1/(np.sqrt(2 * np.pi))) * np.exp(-0.5 * x_vals**2)
plt.plot(x_vals, pdf_vals, 'k', label='PDF')
plt.xlabel('Value')
plt.ylabel('Frequency/Probability')
plt.title('Q5: Histogram with PDF Overlay')
plt.legend()
plt.show()
```
